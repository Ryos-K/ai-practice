# Activation Functions

## Import packages

```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append("..")
import mynn
```

## Prepare activation functions

```python
prelu      = mynn.PReLU()
relu       = mynn.ReLU()
sigmoid    = mynn.Sigmoid()
silu       = mynn.SiLU()
softmax    = mynn.Softmax()
```

## Plot the output of each function for the linear function

```python
coeffs = {
    'A1': 0.5,
    'B1': 2,
    'A2': 1.5,
    'B2': 2,
    'A3': 1.0,
    'B3': 2
}

# Setup the data

START, STOP, NUM = -10, 10, 100
x1 = np.linspace(START, STOP, NUM)
x2 = np.linspace(START, STOP, NUM)
x3 = np.linspace(START, STOP, NUM)
y1 = x1 * coeffs['A1'] + coeffs['B1']
y2 = x2 * coeffs['A2'] + coeffs['B2']
y3 = x3 * coeffs['A3'] + coeffs['B3']

# Plot Original Data
plt.rcParams['figure.figsize'] = [12, 8]
fig = plt.figure()
ax = fig.add_subplot(2, 3, 1)
ax.plot(x1, y1, 'r', label=f'y = {coeffs['A1']} + {coeffs['B1']}')
ax.plot(x2, y2, 'g', label=f'y = {coeffs['A2']} + {coeffs['B2']}')
ax.plot(x3, y3, 'b', label=f'y = {coeffs['A3']} + {coeffs['B3']}')
ax.legend()
ax.set_title('Original Data')

# Plot the output of each activation function
activation_functions: list[mynn.module.Module] = [prelu, relu, sigmoid, silu, softmax]
for i, activation_function in enumerate(activation_functions):
    ax = fig.add_subplot(2, 3, i+1+1)
    output = activation_function.forward(np.concatenate((y1, y2, y3)).reshape(3, -1).T).T
    ax.plot(x1, output[0], 'r', label=f'y = {coeffs['A1']} + {coeffs['B1']}')
    ax.plot(x2, output[1], 'g', label=f'y = {coeffs['A2']} + {coeffs['B2']}')
    ax.plot(x3, output[2], 'b', label=f'y = {coeffs['A3']} + {coeffs['B3']}')
    ax.legend(loc='upper left')
    ax.set_title(activation_function.__class__.__name__)

plt.show()
```

## Plot the output of each function for the sine function

```python
coeffs = {
    'A1': 0.2,
    'B1': 0,
    'A2': 0.3,
    'B2': 0,
    'A3': 0.2,
    'B3': 1
}

# Setup the data

START, STOP, NUM = -10, 10, 100
x1 = np.linspace(START, STOP, NUM)
x2 = np.linspace(START, STOP, NUM)
x3 = np.linspace(START, STOP, NUM)
y1 = np.sin(x1 * coeffs['A1']) + coeffs['B1']
y2 = np.sin(x2 * coeffs['A2']) + coeffs['B2']
y3 = np.sin(x3 * coeffs['A3']) + coeffs['B3']

# Plot Original Data
plt.rcParams['figure.figsize'] = [12, 8]
fig = plt.figure()
ax = fig.add_subplot(2, 3, 1)
ax.plot(x1, y1, 'r', label=f"y = sin({coeffs['A1']}x) + {coeffs['B1']}")
ax.plot(x2, y2, 'g', label=f"y = sin({coeffs['A2']}x) + {coeffs['B2']}")
ax.plot(x3, y3, 'b', label=f"y = sin({coeffs['A3']}x) + {coeffs['B3']}")
ax.legend()
ax.set_title('Original Data')

# Plot the output of each activation function
activation_functions: list[mynn.module.Module] = [prelu, relu, sigmoid, silu, softmax]
for i, activation_function in enumerate(activation_functions):
    ax = fig.add_subplot(2, 3, i+1+1)
    output = activation_function.forward(np.concatenate((y1, y2, y3)).reshape(3, -1).T).T
    ax.plot(x1, output[0], 'r', label=f"y = sin({coeffs['A1']}x + {coeffs['B1']})")
    ax.plot(x2, output[1], 'g', label=f"y = sin({coeffs['A2']}x + {coeffs['B2']})")
    ax.plot(x3, output[2], 'b', label=f"y = sin({coeffs['A3']}x + {coeffs['B3']})")
    ax.legend(loc='upper left')
    ax.set_title(activation_function.__class__.__name__)

plt.show()
```
